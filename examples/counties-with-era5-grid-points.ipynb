```python
import pandas as pd
import geopandas as gpd
import os
import xarray as xr
from datetime import datetime
```

```python
tmp = None
tmp = pd.read_csv("https://zenodo.org/records/7790257/files/maize_countyStats_doyPercentPlanted_2000-2020_DeinesEtAl_vRSE.csv")
```

```python
tmp['GEOID'] = tmp['GEOID'].astype(str)
```

```python
tmp.dtypes
```

```python
countyfile = None
countyfile = gpd.read_file('../data/tl_2022_us_county/tl_2022_us_county.shp')
#countyfile['GEOID'] = countyfile['GEOID'].astype(int)
```

```python
countyfile.head()
```

```python
countyfile.dtypes
```

```python

cdf = None
cdf = tmp.merge(right=countyfile, on="GEOID", how = "left")
```

```python
gpdcounties = gpd.GeoDataFrame(cdf)
```

```python
gpdcounties.crs
```

```python
gpdcounties
```

```python
era5_grid = None
era5_grid = pd.read_csv("https://raw.githubusercontent.com/modern-agronomy/data/main/datasets/era5_grid.csv")
```

```python
era5_grid
```

```python
gdf = gpd.GeoDataFrame(
    era5_grid, geometry=gpd.points_from_xy(era5_grid.longitude, era5_grid.latitude), crs="EPSG:4269"
)
```

```python
gdf.head()
```

```python
gdf.crs
```

```python
joindf = gpdcounties.sjoin(gdf, how = "inner", predicate = "contains" )
```

```python
joindf
```

```python
joindf.columns
```

```python
joindf[['GEOID','county', 'state', 'county_code', 'state_code', 'state_name',
       'crop', 'Year', 'doy_10', 'doy_25', 'doy_50', 'doy_75', 'doy_90', 'latitude',
       'longitude', 'geohash7']].query("Year == 2000")
```
